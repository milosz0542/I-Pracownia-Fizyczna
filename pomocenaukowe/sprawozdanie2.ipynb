# Sprawozdanie 2 - Badanie zderzeń sprężystych - wykresy


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.array([34, 35, 36])
y1 = np.array([29.35, 30.15, 31.45])
y_err1 = np.array([0.77, 0.57, 1.22])

x2 = np.array([34, 35, 36])
y2 = np.array([28.22, 29.05, 31])
y_err2 = np.array([0.4, 0.11, 1.16])

plt.errorbar(x1, y1, yerr=y_err1, fmt='o', label='hstożka = 18cm', alpha=0.5)
plt.errorbar(x2, y2, yerr=y_err2, fmt='o', label='hstożka = 20cm', alpha=0.5)

plt.xlabel('h [cm]')
plt.ylabel('d [cm]')
plt.title('Zależność zasięgu lotu do wysokości zrzutu')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```
