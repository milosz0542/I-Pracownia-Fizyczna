# Wyznaczanie modułu Younga metodą jednostronnego rozciągania

```python
import matplotlib.pyplot as plt
import numpy as np
import linearregression
```

## Wykres zmiany długości drutu w zależności od obciążenia (Regresja liniowa metodą najmniejszych kwadratów)

```python
y = np.array([1, 2, 3, 4, 5, 6])
x = np.array([0.041, 0.124, 0.207, 0.290, 0.373, 0.498]) # Delta L [mm]

# Change milimeters to meters
x = [i / 1000 for i in x]

# Change x in kg to N
y = [i * 9.81 for i in y]

plt.figure(figsize=(18, 12))

# Linear regression
reg = linearregression.LinearRegression()
a, b = reg.fit(x, y)
y_fit = [a * i + b for i in x]
print(f"a: {a} b: {b}")

y_mean = np.mean(y)
x_mean = np.mean(x)
u_a = np.sqrt((1/(len(x)-2)) * sum([(y[i] - y_fit[i])**2 for i in range(len(x))]) / (sum([(x[i] - x_mean)**2 for i in range(len(x))])))
u_b = u_a * np.sqrt(sum([i**2 for i in x]) / len(x))
print(f"u_a: {u_a} u_b: {u_b}")

plt.plot(x, y_fit)
plt.plot(x, y, 'o')
plt.title("Zmiana długości drutu w zależności od obciążenia")
plt.ylabel("Obciążenie [N]")
plt.xlabel("Przyrost długości [m]")
plt.grid(True)
plt.show()
```
