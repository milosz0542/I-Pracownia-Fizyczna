# Chujwieco

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Inicjalizacja danych

```python
alpha_deg = np.arange(0, 75, 5)
alpha = np.tan(np.deg2rad(alpha_deg))

I_1_ros = np.array([0, 1.5, 4.5, 7.5, 10.5, 15.0, 19.5, 24.0, 28.5, 36.0, 42.0, 49.5, 61.5, 76.5, 100.5])
I_1_mal = np.array([0, 1.5, 4.5, 6.0, 9.0, 13.5, 18.0, 21.0, 25.5, 33.0, 39.0, 48.0, 60.0, 78.0, 100.5])

I_2_ros = np.array([0, 1.5, 4.5, 9.0, 13.5, 16.5, 19.5, 22.5, 34.5, 36.0, 40.5, 54.0, 60.0, 78.0, 85.5])
I_2_mal = np.array([0, 1.5, 3.0, 6.0, 10.5, 13.5, 16.5, 22.5, 27.0, 33.0, 37.5, 49.5, 60.0, 75.0, 85.5])

I_1_ros = I_1_ros / 75
I_1_mal = I_1_mal / 75
I_2_ros = I_2_ros / 75
I_2_mal = I_2_mal / 75

print(I_1_ros)
print(I_1_mal)
print(I_2_ros)
print(I_2_mal)
```

## Wykres zależności kąta od prądu dla połączenia pierwszego

```python
plt.figure(figsize=(18, 12))

plt.plot(alpha, I_1_ros, 'o', label='I1 rosnąco')
plt.plot(alpha, I_1_mal, 'o', label='I1 malejąco')
plt.title("Zależność kąta od prądu dla połączenia +->+ ->-")
plt.xlabel("Tangens kąta")
plt.ylabel("Prąd [A]")
plt.legend()
plt.grid(True)
plt.show()
```

## Wykres zależności kąta od prądu dla połączenia pierwszego - średnia

```python
I_1 = (I_1_ros + I_1_mal) / 2
plt.figure(figsize=(18, 12))

plt.plot(alpha, I_1, 'o', label='I1 średnia')
plt.title("Zależność kąta od prądu dla połączenia +->+ ->-")
plt.xlabel("Tangens kąta")
plt.ylabel("Prąd [A]")
plt.legend()
plt.grid(True)
plt.show()
```

## Wykres zależności kąta od prądu dla połączenia pierwszego - regresja liniowa

```python
a_1, b_1 = np.polyfit(alpha, I_1, 1)
plt.figure(figsize=(18, 12))

plt.plot(alpha, a_1 * alpha + b_1, label=f'y={a_1}x+{b_1}', color='blue')
plt.plot(alpha, I_1, 'o', label='I1 średnia', color='blue')
plt.title("Zależność kąta od prądu dla połączenia +->+ ->-")
plt.xlabel("Tangens kąta")
plt.ylabel("Prąd [A]")
plt.legend()
plt.grid(True)
plt.show()
```

## Wykres zależności kąta od prądu dla połączenia drugiego

```python
plt.figure(figsize=(18, 12))

plt.plot(alpha, I_2_ros, 'o', label='I2 rosnąco')
plt.plot(alpha, I_2_mal, 'o', label='I2 malejąco')
plt.title("Zależność kąta od prądu dla połączenia +->- ->+")
plt.xlabel("Tangens kąta")
plt.ylabel("Prąd [A]")
plt.legend()
plt.grid(True)
plt.show()
```

## Wykres zależności kąta od prądu dla połączenia drugiego - średnia

```python
I_2 = (I_2_ros + I_2_mal) / 2
plt.figure(figsize=(18, 12))

plt.plot(alpha, I_2, 'o', label='I2 średnia')
plt.title("Zależność kąta od prądu dla połączenia +->- ->+")
plt.xlabel("Tangens kąta")
plt.ylabel("Prąd [A]")
plt.legend()
plt.grid(True)
plt.show()
```

## Wykres zależności kąta od prądu dla połączenia drugiego - regresja liniowa

```python
a_2, b_2 = np.polyfit(alpha, I_2, 1)
plt.figure(figsize=(18, 12))

plt.plot(alpha, a_2 * alpha + b_2, label=f'y={a_2}x+{b_2}', color='blue')
plt.plot(alpha, I_2, 'o', label='I2 średnia', color='blue')
plt.title("Zależność kąta od prądu dla połączenia +->- ->+")
plt.xlabel("Tangens kąta")
plt.ylabel("Prąd [A]")
plt.legend()
plt.grid(True)
plt.show()
```

## Analiza danych

```python
S_1 = np.sqrt((np.sum(I_1 - (a_1 * alpha + b_1))**2) / (len(I_1) - 2))
S_2 = np.sqrt((np.sum(I_2 - (a_2 * alpha + b_2))**2) / (len(I_2) - 2))

D_1 = len(I_1) * np.sum(alpha**2) - (np.sum(alpha))**2
D_2 = len(I_2) * np.sum(alpha**2) - (np.sum(alpha))**2

u_a_1 = S_1 * np.sqrt(len(I_1) / D_1)
u_a_2 = S_2 * np.sqrt(len(I_2) / D_2)
print(f'Niepewność współczynnika a dla I1: {u_a_1}')
print(f'Niepewność współczynnika a dla I2: {u_a_2}')
```
