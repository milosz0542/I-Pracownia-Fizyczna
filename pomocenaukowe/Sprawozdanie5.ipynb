```python
import numpy as np
import matplotlib.pyplot as plt
import linearregression
```

```python
x = np.array([2, 3, 4, 6, 8, 10])
y_vmaxhmin = np.array([5, 10, 15, 30, 45, 70])
y_vmaxhmax = np.array([10, 15, 20, 25, 35, 45])
y_vavghmin = np.array([0, 0, 0, 5, 5, 10])
y_vavghmax = np.array([10, 10, 10, 15, 20, 25])
y_vminhmin = np.array([0, 0, 0, 5, 5, 10])
y_vminhmax = np.array([10, 10, 15, 25, 35, 45])

plt.figure(figsize=(18, 12))
plt.plot(x, y_vmaxhmin, 'o', label='vmaxhmin')
plt.plot(x, y_vmaxhmax, 'o', label='vmaxhmax')
plt.plot(x, y_vavghmin, 'o', label='vavghmin')
plt.plot(x, y_vavghmax, 'o', label='vavghmax')
plt.plot(x, y_vminhmin, 'o', label='vminhmin')
plt.plot(x, y_vminhmax, 'o', label='vminhmax')
plt.title("Zależność zakrzywienia toru kulki od prędkości i wysokości")
plt.xlabel("r")
plt.ylabel("phi")
plt.legend()
plt.grid(True)
plt.show()
```

```python
a_vmaxhmin, b_vmaxhmin = linearregression.LinearRegression().fit(x, y_vmaxhmin)
a_vmaxhmax, b_vmaxhmax = linearregression.LinearRegression().fit(x, y_vmaxhmax)
a_vavghmin, b_vavghmin = linearregression.LinearRegression().fit(x, y_vavghmin)
a_vavghmax, b_vavghmax = linearregression.LinearRegression().fit(x, y_vavghmax)
a_vminhmin, b_vminhmin = linearregression.LinearRegression().fit(x, y_vminhmin)
a_vminhmax, b_vminhmax = linearregression.LinearRegression().fit(x, y_vminhmax)

plt.figure(figsize=(18, 12))
plt.plot(x, a_vmaxhmin * x + b_vmaxhmin, label=f'vmaxhmin; y = {a_vmaxhmin:.2f}x + {b_vmaxhmin:.2f}')
plt.plot(x, a_vmaxhmax * x + b_vmaxhmax, label=f'vmaxhmax; y = {a_vmaxhmax:.2f}x + {b_vmaxhmax:.2f}')
plt.plot(x, a_vavghmin * x + b_vavghmin, label=f'vavghmin; y = {a_vavghmin:.2f}x + {b_vavghmin:.2f}')
plt.plot(x, a_vavghmax * x + b_vavghmax, label=f'vavghmax; y = {a_vavghmax:.2f}x + {b_vavghmax:.2f}')
plt.plot(x, a_vminhmin * x + b_vminhmin, label=f'vminhmin; y = {a_vminhmin:.2f}x + {b_vminhmin:.2f}')
plt.plot(x, a_vminhmax * x + b_vminhmax, label=f'vminhmax; y = {a_vminhmax:.2f}x + {b_vminhmax:.2f}')
plt.title("Zależność zakrzywienia toru kulki od prędkości i wysokości")
plt.xlabel("r")
plt.ylabel("phi")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Error values
x_err = 0.5
y_err = 2.5

plt.figure(figsize=(18, 12))

# Plot with error bars
plt.errorbar(x, y_vmaxhmin, xerr=x_err, yerr=y_err, fmt='o', label='vmaxhmin')
plt.errorbar(x, y_vmaxhmax, xerr=x_err, yerr=y_err, fmt='o', label='vmaxhmax')
plt.errorbar(x, y_vavghmin, xerr=x_err, yerr=y_err, fmt='o', label='vavghmin')
plt.errorbar(x, y_vavghmax, xerr=x_err, yerr=y_err, fmt='o', label='vavghmax')
plt.errorbar(x, y_vminhmin, xerr=x_err, yerr=y_err, fmt='o', label='vminhmin')
plt.errorbar(x, y_vminhmax, xerr=x_err, yerr=y_err, fmt='o', label='vminhmax')

plt.plot(x, a_vmaxhmin * x + b_vmaxhmin, label=f'vmaxhmin; y = {a_vmaxhmin:.2f}x + {b_vmaxhmin:.2f}')
plt.plot(x, a_vmaxhmax * x + b_vmaxhmax, label=f'vmaxhmax; y = {a_vmaxhmax:.2f}x + {b_vmaxhmax:.2f}')
plt.plot(x, a_vavghmin * x + b_vavghmin, label=f'vavghmin; y = {a_vavghmin:.2f}x + {b_vavghmin:.2f}')
plt.plot(x, a_vavghmax * x + b_vavghmax, label=f'vavghmax; y = {a_vavghmax:.2f}x + {b_vavghmax:.2f}')
plt.plot(x, a_vminhmin * x + b_vminhmin, label=f'vminhmin; y = {a_vminhmin:.2f}x + {b_vminhmin:.2f}')
plt.plot(x, a_vminhmax * x + b_vminhmax, label=f'vminhmax; y = {a_vminhmax:.2f}x + {b_vminhmax:.2f}')

plt.title("Zależność zakrzywienia toru kulki od prędkości i wysokości")
plt.xlabel("r")
plt.ylabel("phi")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(18, 12))

# Plot with error bars
plt.errorbar(x, y_vmaxhmin, xerr=x_err, yerr=y_err, fmt='o', label='vmaxhmin', alpha=0.3, capsize=5)
plt.errorbar(x, y_vmaxhmax, xerr=x_err, yerr=y_err, fmt='o', label='vmaxhmax', alpha=0.3, capsize=5)
plt.errorbar(x, y_vavghmin, xerr=x_err, yerr=y_err, fmt='o', label='vavghmin', alpha=0.3, capsize=5)
plt.errorbar(x, y_vavghmax, xerr=x_err, yerr=y_err, fmt='o', label='vavghmax', alpha=0.3, capsize=5)
plt.errorbar(x, y_vminhmin, xerr=x_err, yerr=y_err, fmt='o', label='vminhmin', alpha=0.3, capsize=5)
plt.errorbar(x, y_vminhmax, xerr=x_err, yerr=y_err, fmt='o', label='vminhmax', alpha=0.3, capsize=5)

plt.title("Zależność zakrzywienia toru kulki od prędkości i wysokości")
plt.xlabel("r")
plt.ylabel("phi")
plt.legend()
plt.grid(True)
plt.show()
```

## Siła Coriolisa


```python
y_coriolis = np.array([0.359, 0.360, 0.362, 0.367, 0.374, 0.382])

polyfit_coriolis = np.polyfit(x, y_coriolis, 2)
poly_coriolis = np.poly1d(polyfit_coriolis)

# Generate polynomial curve
x_fit = np.linspace(0, 10, 100)
y_fit = poly_coriolis(x_fit)

plt.figure(figsize=(18, 12))
plt.plot(x, y_coriolis, 'o', label='Coriolis')
plt.plot(x_fit, y_fit, label=f'Coriolis')
plt.title("Zależność siły Coriolisa od promienia R")
plt.xlabel("r")
plt.ylabel("F")
plt.legend()
plt.grid(True)
plt.show()
```
